```python
# here n is a user defined input
n=int(input(''))
x=0
for i in range(0,n+1):
    
    a=df1['Average Price'][i]
    x=x+a
    
b=n+1
closing_price_0=df1['Close Price'][0]
closing_price_nth=df1['Close Price'][n]
if(closing_price_nth>closing_price_0):
    profit_percentage=((closing_price_nth-closing_price_0)/closing_price_0)*100
    print('profit of last {} days is {}%'.format(n,profit_percentage))
else:
    loss_percentage=((closing_price_nth-closing_price_0)/closing_price_0)*100
    print('Loss of last {} days is {}%'.format(n,loss_percentage))
    
# print the average value over last n days
print('The average price of last {} days is {}'.format(n,x/b))

# it will also calculate the average price and profit/loss percentage of last 1 week=7 days,2weeks=14 days,1month=30days
# 2 months=60days,3 months=90 days,6 months=180 days,i year=365 days

```

```python
# pct_change function of pandas get the percentage change between two days 
perc=df1[['Close Price']].pct_change()
# we multiply by 100 to get the change in percentage
perc=perc*100
```

```python
# the first value is NAN because the pct_change function compares two values thats why for row 0 there is no comparable data
# so we relace the NAN value by 0 we use fillna() function of pandas and replace by 0
df1['Day_perc_change']=perc.fillna(0)
```

```python
df1
```

```python
# here the dtype of Day_perc_change column in object so we change to float64 of pandas by function astype
df1[['Day_perc_change']].astype('float64')
```

```python
df1['Day_perc_change']=df1['Day_perc_change'].astype('float64')
```

```python
df1['Day_perc_change'].dtypes
```

```python
df1['Trend']='NAN'
```

```python
df1
```

```python
for i in range(0,494):
    value=df1[['Day_perc_change']].iloc[i]
    real_value=value['Day_perc_change']
    if(real_value>-0.5 and real_value<=0.5):
        df1['Trend'][i]='Slight or No change'
    elif(real_value>0.5 and real_value<=1):
        df1['Trend'][i]='Slight positive'
    elif(real_value>-1 and real_value<-0.5):
        df1['Trend'][i]='Slight negative'
    elif(real_value>1 and real_value<=3):
        df1['Trend'][i]='positive'
    elif(real_value>-3 and real_value<-1):
        df1['Trend'][i]='negative'
    elif(real_value>3 and real_value<=7):
        df1['Trend'][i]='Among top gainers'
    elif(real_value>-7 and real_value<-3):
        df1['Trend'][i]='Among top losers'
    elif(real_value>7):
        df1['Trend'][i]='Bull run'
    elif(real_value<=-7):
        df1['Trend'][i]='Bear drop'
        
    
```
